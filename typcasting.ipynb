```python
# int to float,boolean,string,complex

#int to float

a = 100           #int----> 100
print(a,type(a))
b= float(a)       #int to float--->output---> 100.0
print(b,type(b))
```

```python
#int to boolean(True)

a = 100            #int---->100
print(a,type(a))
c = bool(a)        #int to bool----> output---> True
print(c,type(c))   #The result is False only when the value is zero.
```

```python
#int to boolean(False)

a1 = 00             #int---->00
print(a,type(a))
c1 = bool(a1)        #int to bool----> output---> False
print(c1,type(c1))
```

```python
#int to string

a = 100            #int value ----> 100
print(a,type(a))
d = str(a)         #int to str----> output-----> "100"
print(d,type(d))
```

```python
#int to complex

a = 100            #int value ----> 100
print(a,type(a))
e = complex(a)     #int to complex----->output------>(100+0j)
print(e,type(e))
```
