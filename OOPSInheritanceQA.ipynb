```python
#1. Create a Vehicle class with attributes brand and speed. Create a child class Car that inherits from Vehicle and adds fuel_type.

class Vehicle:
    def __init__(self,brand,speed):
        self.brand = brand
        self.speed = speed
        print("Brand :",self.brand)
        print("Speed :", self.speed)
        
class Car(Vehicle):
    
    def __init__(self,brand,speed,fuel_type):
        Vehicle.__init__(self,brand,speed)
        self.fuel_type = fuel_type
        print("Fuel Type :",fuel_type)
        
car1 = Car("BMW","300km/h","Petrol")
```

```python
#3. Create a Shape class with a method area(). Create subclasses Rectangle and Circle that calculate their respective areas.

class Shape:
  def area(self):
    print("Area of Shape")

class Rectangle(Shape):
  def area(self,length,breadth):
    self.length = length
    self.breadth = breadth
    Area_of_Rectangle = self.length * self.breadth
    print("Area of Rectangle is :",Area_of_Rectangle)

class Circle(Shape):
  def area(self,radius):
    self.radius = radius
    Area_of_Circle = 3.14 * self.radius * self.radius
    print("Area of Circle is :",Area_of_Circle)


R1 = Rectangle()
R1.area(25,35)
C1 = Circle()
C1.area(7)      
```

```python
#5. Create a BankAccount class with attributes account_number and balance. Create a subclass SavingsAccount with an additional interest rate.

class BankAccount:
    def __init__(self,account_number,balance):
        self.account_number = account_number
        self.balance = balance
        print("Account number :",self.account_number)
        print("Balance :",self.balance)

class SavingsAccount(BankAccount):
    def __init__(self,account_number,balance,interest_rate):
        super().__init__(account_number,balance)
        self.interest_rate = interest_rate
        print("Interest Rate :",self.interest_rate)
    

A1 = SavingsAccount(1234567,250000,"5%")
```

```python
#8. Create a Vehicle class → Car class → ElectricCar class with additional battery capacity.

class Vehicle:
    def __init__(self,brand):
        self.brand = brand
        print("Brand :",self.brand)

class Car(Vehicle):
    def __init__(self,brand):
        super().__init__(brand)

class ElectricCar(Car):
    def __init__(self,brand,battery_capacity):
        super().__init__(brand)
        self.battery_capacity = battery_capacity
        print("Battery Capacity :", battery_capacity)

E1 = ElectricCar("Tata Nexon EV", "82 kWh")
```

```python
#9. Create a Person class → Employee class → Developer class with programming language specialization.

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        print("Name :",self.name)
        print("Age :",self.age)

class Employee(Person):
    def __init__(self,name,age,employee_ID,salary):
        super().__init__(name,age)
        self.employee_ID = employee_ID
        self.salary = salary
        print("Employee ID :",self.employee_ID)
        print("Salary :",self.salary)

class Developer(Employee):
    def __init__(self,name,age,employee_ID,salary,programming_language):
        super().__init__(name,age,employee_ID,salary)
        self.programming_language = programming_language
        print("Programming Language Specialization :",programming_language)


D1 = Developer("Sachin",34,"AB12345",55000,"JAVA")
        
```

```python
#10. Create a class hierarchy for University → Department → Course.

class University:
    def __init__(self,name):
        self.name = name
        print("University :",self.name)

class Department(University):
    def __init__(self,name,department_name):
        super().__init__(name)
        self.department_name = department_name
        print("Department :",self.department_name)

class Course(Department):
    def __init__(self,name,department_name,course_name):
        super().__init__(name,department_name)
        self.course_name = course_name
        print("Course :",self.course_name)

U1 = Course("DAVV University","Management","MBA")
    
```
