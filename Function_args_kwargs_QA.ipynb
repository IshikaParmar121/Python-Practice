```python
# 1. Grade Evaluation System
#    Write a function calculate_grade(marks).
#    The function should:

# Accept marks between 0 and 100.
# Return "A" if marks ≥ 90
# Return "B" if marks ≥ 75 and < 90
# Return "C" if marks ≥ 60 and < 75
# Return "D" if marks ≥ 40 and < 60
# Return "Fail" if marks < 40
# If marks are negative or greater than 100, return "Invalid Marks"

def calculate_grade(marks):
    if marks < 0 or marks > 100:
        return "Invalid Marks"

    if marks >= 90:
        return "A"
    elif marks >= 75 and marks < 90:
        return"B"
    elif marks >= 60 and marks < 75:
        return"C"
    elif marks >= 40 and marks < 60:
        return"D"
    else:
        return"Fail"
marks = int (input("Marks : "))
print("The grade is : ",calculate_grade(marks))
```

```python
# 2. Leap Year Checker
#    Write a function is_leap_year(year).
#    The function should:

# Return "Leap Year" if the year is divisible by 400
# OR if the year is divisible by 4 but not divisible by 100
# Otherwise return "Not a Leap Year"

def is_leap_year(year):
    if year % 400== 0 or year % 4==0 and year % 100!=0:
        return"Leap Year"
    else:
        return"Not a Leap Year"
year = int(input("Year : "))
print("The Year is : ",is_leap_year(year))
```

```python
# 3. Salary Tax Calculator
#    Write a function calculate_tax(salary).
#    The function should:

# Return 0 if salary ≤ 250000
# Apply 5% tax if salary is between 250001 and 500000
# Apply 10% tax if salary is between 500001 and 1000000
# Apply 20% tax if salary is above 1000000
# Return total tax amount
# If salary is negative, return "Invalid Salary"

def calculate_tax(salary):
    if salary < 0:
        tax = "Invalid Salary"
    elif salary <= 250000:
        tax = 0
    elif salary <= 500000:
        tax = salary*0.05
    elif salary <= 1000000:
        tax = salary*0.10
    else:
        tax = salary*0.20
    return f"Salary {tax}"
        
calculate_tax(1100000)
```

```python
# 5. Triangle Type Identifier
#    Write a function triangle_type(a, b, c).
#    The function should:

# First check if the triangle is valid (sum of any two sides must be greater than the third side)
# If not valid, return "Not a Valid Triangle"
# If valid and all sides equal, return "Equilateral"
# If valid and two sides equal, return "Isosceles"
# If all sides different, return "Scalene"

def triangle_type(a, b, c):
    if a + b < c and a + c < b and b + c < a:
        return"Not a Valid Triangle"
    
    if a == b == c:
        return "Equilateral"
    elif a==b or a==c or b==c:
        return "Isosceles"
    else:
        return"Scalene"
a = int(input("Value of a : "))
b = int(input("Value of b : "))
c = int(input("Value of c : "))
print("The triagle type is : ",triangle_type(a, b, c))   
```

```python
# 6. Electricity Bill Generator
#    Write a function electricity_bill(units).
#    Conditions:

# First 100 units: ₹5 per unit
# Next 100 units: ₹8 per unit
# Above 200 units: ₹10 per unit
# If units are negative, return "Invalid Input"
# Return total bill amount

def electricity_bill(units):
    if units < 0:
        bill="Invalid Input"
    elif units<= 100:
        bill= units*5
    elif units<=200:
        bill= (100*5) + (units- 100)*8 
    else:
        bill=(100*5)+(100*8)+(units- 200)*10
    return bill
    
units = int(input("The total units are : "))    
print("Total bill amount : ",electricity_bill(units))       
```

```python
# 7. Number Category Analyzer
#    Write a function analyze_number(num).
#    The function should:

# Check whether the number is positive, negative, or zero
# Also check whether it is even or odd
# Return result in format:
#   "Positive Even"
#   "Positive Odd"
#   "Negative Even"
#   "Negative Odd"
#   "Zero"

def analyze_number(num):
    if num > 0 and num % 2==0:
        return "Number is Positive Even"
    elif num > 0 and num % 2!= 0:
        return "Number is Positive Odd"
    elif num < 0 and num % 2==0:
        return "Number is Negative Even"
    elif num < 0 and num % 2!=0:
        return "Number is Negative Odd"
    else:
        return "Number is Zero"
num = int(input("Number : "))
analyze_number(num)
```

```python
# 1. Sum of Unlimited Numbers
# Write a function calculate_total(*args).
# Accept any number of numeric arguments
# Return the sum of all numbers
# If no arguments are passed, return 0

def total(*args):
    if args == ():
        return 0
        
    num = 0
    for i in args:
        num = num + i
    return num
print("Sum of all numbers : ",total(1,2,30))
print(total())
```

```python
# 2. Find Maximum Using *args
# Write a function find_max(*args). The function should:
# Accept any number of numbers
# Return the maximum value
# If no arguments are provided, return "No Values Provided"

def find_max(*args):
    if args== ():
        return "No Values Provided"
    maximum = args[0]
    for num in args:
        if num > maximum:
            maximum=num
    return maximum
        
print("The maximum number is : ",find_max(1,89,3))
print(find_max())
```

```python
# 3. Even and Odd Counter
# Write a function count_even_odd(*args). The function should:
# Accept any number of integers
# Return a tuple containing (even_count, odd_count)

def count_even_odd(*args):
    even_count = 0
    odd_count = 0
    for number in args:
        if number % 2 == 0:
            even_count = even_count + 1
        else:
            odd_count = odd_count + 1
    return (even_count,odd_count)
print("The even and odd count is:",count_even_odd(1,2,3,4,7,8,78,89,90))
```

```python
# 4. Product of Numbers with Initial Value
# Write a function multiply_numbers(initial_value, *args). The function should:
# Multiply initial_value with all additional numbers
# Return the final result
# If no extra arguments are given, return initial_value

def multiply_numbers(initial_value,*args):
    
    for numbers in args:
        initial_value = initial_value * numbers
    return initial_value
    
print("Multiplication of numbers :",multiply_numbers(2,3,4,7))
print("Initial Value :",multiply_numbers(9))
```

```python
# 5. Student Details Printer
# Write a function student_info(**kwargs). The function should:
# Accept any number of keyword arguments
# Print each key and value in the format: key : value
# If no data is passed, print "No Details Provided"



def student_info(**kwargs):
    if kwargs=={}:
        print("No Details Provided")
    for key,value in kwargs.items():
        print(key,":",value)

student_info(Name = "Nishi Sahu",Age = 23)        
```

```python
# 6. Employee Salary Calculator
# Write a function calculate_salary(**kwargs). The function should expect:
# basic_salary (mandatory key)
# bonus (optional, default 0)
# deductions (optional, default 0)
# Return final salary after adding bonus and subtracting deductions.
# If basic_salary is missing, return "Basic Salary Required"


def calculate_salary(**kwargs):
    if "basic_salary" not in kwargs:
        return "Basic Salary Required"

    basic_salary = kwargs["basic_salary"]

    if "bonus" in kwargs:
        bonus = kwargs["bonus"]
    else:
        bonus = 0

    if "deduction" in kwargs:
        deduction = kwargs["deduction"]
    else:
        deduction = 0

    final_salary = basic_salary + bonus - deduction
    return final_salary

print("Final Salary :",calculate_salary(basic_salary = 39000,bonus = 2000,deduction = 1000)) 
print("Final Salary :",calculate_salary(basic_salary = 60000,bonus = 5000))
print("FinalSalary :",calculate_salary(basic_salary = 70000,deduction = 4000))
print("Final Salary :",calculate_salary(basic_salary = 50000))
print(calculate_salary(bonus = 2000))
```

```python
# 7. Keyword Validation
# Write a function validate_user(**kwargs). The function should:
# Check if both "username" and "password" keys exist
# If both exist, return "Valid Data"
# Otherwise return "Missing Required Fields"

def validate_user(**kwargs):
    
    if "username" in kwargs and "password" in kwargs:
        return "Valid Data"
    else:
        return "Missing Required Fields"

print(validate_user(username = "Pihu",password = 1234))
print(validate_user(username = "Kia"))
print(validate_user(password = 98765)) 
print(validate_user())
```

```python
#8. Combining *args and **kwargs
# Write a function order_summary(*args, **kwargs). The function should:
# Treat *args as item prices
# Treat **kwargs as additional charges (like tax=, delivery=)
# Return total bill amount

def order_summary(*args,**kwargs):
    item_price = sum(args)
    
    if "tax" in kwargs:
        tax = kwargs["tax"]
    else:
        tax = 0

    if "delivery_charges" in kwargs:
        delivery_charges = kwargs["delivery_charges"]
    else:
        delivery_charges = 0
        
    total_bill = item_price + tax + delivery_charges
    return total_bill

print("Total Bill Amount:",order_summary(250,tax = 50, delivery_charges = 35))
print("Total Bill Amount:",order_summary(200,300,300,400))
print("Total Bill Amount:",order_summary(200,500,700,tax = 500))
print("Total Bill Amount:",order_summary(200,500,600,300,delivery_charges = 1000)) 
```

```python
# 9. Filter Positive Numbers
# Write a function filter_positive(*args). The function should:
# Return a list of only positive numbers from given arguments
# If no positive numbers exist, return an empty list  

def filter_positive(*args):
    positive_numbers = []
    
    for numbers in args:
        if numbers > 0:
            positive_numbers.append(numbers)
    return positive_numbers

print(filter_positive(2,3,4,5,-6,-7,-8))
print(filter_positive(-9,-8,-7))
```

```python
# 10. Flexible Greeting System
# Write a function greet_user(*args, **kwargs). The function should:
# Use first positional argument as name (if provided)
# If no positional argument, check for name in kwargs
# If name not provided anywhere, return "Guest"
# Optionally use title from kwargs (like title="Dr.")

def greet_user(*args,**kwargs):
    if len(args) > 0:
        name = args[0]
    elif "name" in kwargs:
        name = kwargs["name"]
    else:
        name = "Guest"

    if "title" in kwargs:
        title = kwargs["title"]
        return title + " " + name
    else:
        return name
        
print(greet_user("Ayushi"))
print(greet_user(name="Raj"))
print(greet_user(title="Dr.", name="Sinha"))
print(greet_user())
```
