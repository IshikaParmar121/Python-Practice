```python
# Class : Company,attributes = name,sal,dept.....
# Sal() print salary
# func() named = bonus==> salary 5% add,

class Company:
    def __init__(self,name,sal,dept):
        self.name = name
        self.sal = sal
        self.dept = dept
        print("Good morning everyone!\nA little extra happiness today because it’s **Salary Day!\nHope everyone has a productive and cheerful day ahead.")
        print()
        
    def print_salary(self):
        print("The Salary of",self.name,"who is from",self.dept,"Department, is :",self.sal)

    def func(self):
        print("The Salary with 5% Bonus is :",self.sal + self.sal * 0.05)
        
employee1 = Company("Anisha",25000,"Management")
employee1.print_salary()
employee1.func()

print()

employee2 = Company("Megha",67000,"Finance")
employee2.print_salary()
employee2.func()
```

```python
#5. Write a class BankAccount with attributes account_holder and balance. Add a method deposit(amount) that increases the balance.

class BankAccount:
    def __init__(self,account_holder,balance):
        self.account_holder = account_holder
        self.balance = balance
    def deposit(self,amount):
        self.amount = amount
        print("Account Holder :",self.account_holder,", Balance :",self.balance)
        print("Deposit Amount :",self.amount)
        self.balance = self.balance + amount
        print("Increased Balance = Balance + Deposit Amount :",self.balance)

A1 = BankAccount ("'Jiya Gupta'",20000)
A1.deposit(20000)    
```

```python
#7. Write a class Rectangle that takes length and width in init and has a method area() that returns the area.

class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def area(self):
        area = self.length * self.width
        return f"Area of Rectangle is : {area}"
a1 = Rectangle(9,3)
print(a1.area())
```

```python
#9. Write a class Calculator with methods add(a, b) and show_result(result) that prints the result.

class Calculator:
    def add(self,a,b):
        self.a = a
        self.b = b
        result = a + b
        return result
    def show_result(self,result):
        print("The addition of",self.a,"and",self.b,"is:",result)
A1 = Calculator()
result = A1.add(48,67)
A1.show_result(result)    
```

```python
#10. Write a class Counter that counts how many objects have been created using a class variable.

class Counter:
    count = 0
    def __init__(self):
        Counter.count = Counter.count + 1
    print("The total number of objects :",Counter.count)
c1 = Counter()
c2 = Counter()
c3 = Counter()    
```
