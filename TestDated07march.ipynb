```python
#1. Write a function sum_numbers(*args) that returns the sum of all numbers passed as arguments.

def sum_numbers(*args):
    total = 0
    for i in args:
        total = total + i
    print("Sum of numbers passed as arguments : ",total)
sum_numbers(1,2,3,4,5)
```

```python
#2. Write a function count_args(*args) that returns the total number of arguments passed to the function.
def count_args(*args):
    count = len(args)
    print("Total number of arguments passed to the function is : ",count)
count_args(1,2,3,4,5,6,7)
```

```python
#3. Write a function find_max(*args) that returns the largest value among all the arguments.

def find_max(*args):
    max_num = 0
    for i in args:
        if i > max_num:
            max_num = i
    return f"The Largest value among all the arguments is : {max_num}"
find_max(1,2,3,34,4,5,6)
```

```python
#4. Write a function print_items(*args) that prints each argument on a separate line.

def print_items(*args):
    for num in args:
        print(num)
print_items(1,2,3,4,4,56,7,8)    
```

```python
#5. Write a function multiply_numbers(*args) that returns the product of all the numbers passed as arguments.

def multiply_numbers(*args):
    product = 1
    for numbers in args:
        product = product * numbers
    return f"Product of all numbers passed as arguments : {product}"
multiply_numbers(1,2,3,4,5,10)
```

```python
#6. Write a function greet_user(**kwargs) that prints "Hello <name>" if the key name is provided. If name is not provided, print "Hello Guest"

def greet_user(**kwargs):
    if "name" in kwargs:
        print("Hello",kwargs["name"])
    else:
        print("Hello Guest")
print(greet_user(name = "Megha"))
print(greet_user())
```

```python
#10. Write a function get_even_numbers(*args) that returns a list containing only the even numbers from the arguments.

def get_even_numbers(*args):
    even_list = []
    for numbers in args:
        if numbers % 2 == 0:
            even_list.append(numbers)
    return f"The even numbers from the arguments :{ even_list}"
            
get_even_numbers(1,5,6,4,8,80,78,68)
```

```python
#13. Write a function combine_arguments(*args, **kwargs) that prints all positional arguments first and then prints all keyword arguments.

def combine_arguments(*args, **kwargs):
    print("Positional Arguments:")
    for i in args:
      print(i)
    print("Keyword Arguments:")
    for key,value in kwargs.items():
        print(key,":",value)
combine_arguments("Amit",1,2,3,4,Name= "Disha",Surname = "Verma",Age= 12,City = "Indore")
```

```python
#14. Write a function process_order(*items, **details) that prints:
# The list of items ordered
# Customer name
# Payment method.

def process_order(*items, **details):
    print("List of Items ordered:")
    for i in items:
        print(i)
    print("Customer Details:")
    print("Customer Name:",details["Name"])
    print("Payment Method:",details["Payment_Method"])
    
process_order("Shampoo","Soap","Fruits","Books",Name = "Disha",Payment_Method = "Online")
```

```python
#15. Write a function calculate_square(n) that contains a nested function square() which returns the square of n.

def calculate_square(n):
    def square():
        return n * n
    print(square())
calculate_square(6)   
```

```python
#17. Write a function calculator(a, b, operation) that contains nested functions add(), subtract(), and multiply(). 
#Call the correct nested function based on the value of operation.

def calculator(a, b, operation):
    def add():
        if operation == "+":
            print("The Addition is :",a + b)
    add()
    def substract():
        if operation == "-":
            print("The Subtraction is :",a-b)
    substract()
    def multiply():
        if operation == "*":
            print("The Multiplication is :",a*b)
    multiply()

a = int(input("First Number is  :"))
b = int(input("Second Number is :"))
operation = input("Operation is     :")

            
calculator(a,b,operation)
```

```python
#18. Write a function login(username, password) that contains a nested function validate().
# The validate() function should check if:
# username = "admin"
# password = "1234"
# Return "Login Successful" if correct, otherwise return "Invalid Credentials".

def login(username, password):
    def validate():
        if username == "admin" and password == "1234":
            return "Login Successful"
        else:
            return "Invalid Credentials"
    print(validate())

print(login("adin","1235"))
print(login("admin","1234"))
```

```python
#19. Write a function outer_function(x) that contains a nested function inner_function(y) which returns the sum of x and y.
# The outer function should return the inner function.

def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function
new = outer_function(9)
print(new(8))    
```

```python
#20. Write a function calculate_area(length, width) that contains a nested function area() which returns the area of the rectangle.

def calculate_area(length, width):
    def area():
        area_of_rectangle = length * width
        return f"The area of rectangle is :{ area_of_rectangle}"
    return area()
calculate_area(3,9)
```
